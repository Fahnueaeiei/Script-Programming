```python
# --- Lab 3.1 Part 1: Single Number Multiplication Table ---
n = int(input("Enter a number for its multiplication table (e.g., 7): "))
for i in range(1, 11):
    print(f"{n} * {i} = {n*i}")

print()  # spacer
```

```python
# --- Lab 3.1 Part 2: Full 12x12 Multiplication Grid ---
print("12x12 Multiplication Table:")
for i in range(1, 13):
    for j in range(1, 13):
        print(f"{i*j:4}", end=" ")  # Print each product with a width of 4 characters
    print()  # Move to the next line after each row

print()  # spacer
```

## Activity 3.1: Debugging Infinite Loops & Loop Logic Errors
แนวทาง: หา bug ในแต่ละ snippet ด้านล่าง อธิบายว่าทำไมถึงพัง แล้วแก้ไขให้ทำงานถูกต้อง

**Snippet A: Infinite Loop**

โค้ดเดิม:
counter = 0
while counter < 5:
    print(counter)
    # Missing counter increment!

สาเหตุ: ไม่มีการอัปเดตค่า counter ในลูป ทำให้เงื่อนไข counter < 5 เป็น True ตลอดไป เกิด infinite loop
วิธีแก้: เพิ่ม counter += 1 เข้าไปในลูป

**Snippet B: Incorrect Range**

โค้ดเดิม:
for i in range(1, 5):  # Goal: print 1, 2, 3, 4, 5
    print(i)

สาเหตุ: range(1, 5) ให้ค่า 1, 2, 3, 4 เท่านั้น (ไม่รวม stop) แต่ต้องการให้พิมพ์ถึง 5
วิธีแก้: เปลี่ยน stop จาก 5 เป็น 6 -> range(1, 6)

**Snippet C: Misplaced Break**

โค้ดเดิม:
for char in "hello":
    if char == 'l':
        print("Found 'l', stopping!")
        break
    print(char)

สาเหตุ: print(char) อยู่หลังเงื่อนไข if ทำให้ตัวอักษรที่ทำให้ break (l) ไม่ถูกพิมพ์ออกมาก่อน
วิธีแก้: ย้าย print(char) มาไว้ก่อนเช็คเงื่อนไข if เพื่อพิมพ์ตัวอักษรนั้นก่อน break

```python
# --- Lab 3.2 Part 1: Countdown Timer ---
start = int(input("Enter starting number for countdown: "))
while start >= 0:
    print(f"Countdown: {start}")
    start -= 1
print("Blast off!")
```

```python
# --- Lab 3.2 Part 2: Simple Guessing Game ---
secret_number = 42
attempts = 0

while attempts < 5:
    guess = int(input("Guess the number: "))
    attempts += 1

    if guess < secret_number:
        print(f"{guess} is too low! Try again.")
    elif guess > secret_number:
        print(f"{guess} is too high! Try again.")
    else:
        print("Congratulations! You guessed it!")
        break
else:
    print(f"Sorry, you've used all 5 attempts. The number was {secret_number}.")
```
