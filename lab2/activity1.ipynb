```python
# Block A: Evaluation Check
temp = 25
is_raining = True
if temp > 20 and not is_raining:
    print("Perfect day for a picnic!")
elif temp <= 20 or is_raining:
    print("Stay indoors or bring an umbrella.")
else:
    print("Unsure about the weather.")

# คำตอบ Block A:
# Step 1: if temp > 20 and not is_raining
#   -> 25 > 20 = True
#   -> not True = False
#   -> True and False = False  ไม่เข้าเงื่อนไขนี้
#
# Step 2: elif temp <= 20 or is_raining
#   -> 25 <= 20 = False
#   -> is_raining = True
#   -> False or True = True  เข้าเงื่อนไขนี้
#
# Output: "Stay indoors or bring an umbrella."


# Block B: Vectorized Filter
phrase = "Python 2026 Engine"
consonants = {c.lower() for c in phrase if c.isalpha() and c.lower() not in 'aeiou'}
print(f"Unique consonants: {consonants}")

# คำตอบ Block B:
# 1. c.isalpha() -> กรองเฉพาะตัวอักษร ตัดช่องว่างและเลข "2026" ออก
#    เหลือ: P y t h o n E n g i n e
#
# 2. c.lower() -> แปลงทุกตัวเป็นตัวเล็ก
#    เหลือ: p y t h o n e n g i n e
#
# 3. ตัดสระ (a,e,i,o,u) ออก -> เหลือแต่พยัญชนะ
#    p, y, t, h, n, g, n, n
#
# 4. ใช้ {} (set) -> เก็บแบบไม่ซ้ำ ไม่เรียงลำดับตายตัว
#
# Output: Unique consonants: {'p', 't', 'y', 'h', 'n', 'g'}
# (ลำดับใน set อาจสลับกันได้ในแต่ละครั้งที่รัน)
```
